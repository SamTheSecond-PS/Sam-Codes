```python
from collections import deque

dat =  deque()
dat.extend([1,2,3]) #adds numbers to a deque obj
dat.appendleft(2) #adds a element to the front
dat.rotate(1) #puts the last element to front
dat.rotate(-2) #puts the front elements th back
dat.append(4)
dat.pop()
dat.pop()
dat.popleft() #removes element from front
print(dat)
```
