```python
from itertools import cycle

prgn_langs = ["overprotective crab", "Flamethrower to kid dealer", "Flash++", "itch"]
c = cycle(prgn_langs)

for _ in range(6):
    print((next(c)))

'''Repeats all elements in a list.'''
```

```python
from itertools import count

for i in count(1):
    print(i)
    if i > 20:
        break

'''Keeps counting from _ endlessly and doesnt
stop unless manually broken.'''
```

```python
from itertools import permutations, combinations

for p in permutations(["a", "b", "c"], 2):
    print(p)

for c in combinations(['a', 'n', 'n'], 2):
    print(c)

'''Returns all possible outcomes of elements in a list
where x elements are in 1 outcome. --perm'''

'''Returns all possible combinations with no duplicates. --comb'''
```

```python
from itertools import repeat

for x in repeat("hello", 2):
    print(x)

'''Repeats 1 value x times. If no second arg it 
repeats inf.'''
```
