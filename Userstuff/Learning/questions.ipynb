```python
'''QUESTION 1'''

s = "hello guys! This code Q"
def find_chars(lst: list, s: str) -> dict:

    ndict = {}
    for x in lst:
        if x < len(s):
           ndict[x] = s[x]
        else:
            pass
    return ndict

s = find_chars([1 , 2, 5, 7], s)
print(s)
```

```python
'''QUESTION 2'''

lst = [1, 2, 4, 6, 9, 0]

def sumlst(lt: list) -> int:
    total = 0
    for x in lt:
        total += x if x % 2 == 0 else 0
    return total
sumlst(lst)
```
