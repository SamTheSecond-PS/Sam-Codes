```python
def like_system():
    likes = 1

    def like():
        nonlocal likes
        likes += 1
        return likes
    
    return like

btn = like_system()
btn()
btn()
print(btn())
```

```python
def outer(x):
    def inner(y):
        return x + y
    return inner

addto5 = outer(5) #Makes a object for the outer func
print(addto5(2)) #The outer func uses its inner func
#here no need for calling the inner func name
```
